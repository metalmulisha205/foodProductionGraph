```python
from plotly.offline import init_notebook_mode, iplot, plot, download_plotlyjs
init_notebook_mode(connected=True)
import plotly.graph_objs as go
import plotly.offline as py
#offline.init_notebook_mode()
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('FAO.csv', encoding='latin1')
item = "Wheat and products"
year = "2013"
df.head()
```

```python
df2 = df[df.Item == item]
df2 = df2[df2["Element"] == 'Food']
df2



    
```

```python
data = [dict (
        type = 'choropleth',
        locations = df2['Area Abbreviation'],
        z = df2['Y' + year],
        text = df2['Area'],
        colorscale = 'Rainbow',
        autocolorscale = False,
        reversescale = True,
        marker = dict(
            line = dict(
                color = 'rgb(180,180,180)',
                width = 0.5
            )),
        colorbar = dict(
            autotick = False,
            tickprefix = '',
            title = 'Production<br>1000 tonnes'),
        
        ) ]

layout = dict(
    title = 'Showing: ' + item,
    geo = dict(
        showframe = False,
        showcoastlines = False,
        projection = dict(
            type = 'Mercator'
        )
    )
)    

fig = dict(data=data, layout=layout)
```

```python
graph = py.iplot(fig, validate=False)
```
